```python
from utils.database_entry import data_entry

# Install kafka-python
#!pip install kafka-python

from kafka import KafkaConsumer
import json

kafka_config = open('../../config/kafka_config.json')
kafka_config = json.load(kafka_config)
consumer_group_id = kafka_config["consumer_group_id"]

db_config = open('../../config/database.json')
db_config = json.load(db_config)


consumer = KafkaConsumer(kafka_config["kafka_topic"], bootstrap_servers=kafka_config["bootstrap_servers_IP_port"], value_deserializer=lambda x: json.loads(x.decode('utf-8')))

for msg in consumer:
#     print(msg.value)
    data = msg.value
    print(1)
    try:
        data_entry(db_config["pg_password"],"us_data",data)
    except Exception as e:
        print(f"error handling in Data Entry: error={e}")
```
